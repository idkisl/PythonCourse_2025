### Напишем легенду

```python
names = ["Василий", "Артем", "Иван", "Зинаида", "Ваня"]
marks = [5, 4, 7, 6, 1]

plt.scatter(names, marks)
plt.title("Оценки студентов МФТИ", fontsize=20)
plt.xlabel("Студент")
plt.ylabel("Оценка")
plt.show()
```

```python
names = ["Василий", "Артем", "Иван", "Зинаида", "Ваня"]
marks = [5, 4, 7, 6, 1]

plt.scatter(names, marks, color="red", marker="*")
plt.title("Оценки студентов МФТИ", fontsize=20)
plt.xlabel("Студент")
plt.ylabel("Оценка")
plt.show()
```

```python
t = np.linspace(0, 10, 100)
v = 9.8 * t
plt.plot(t, v)
plt.title("Зависимость скорости от времени", fontsize=20)
plt.xlabel("Время (с)")
plt.ylabel("Скорость (м/с)")
plt.show()
```

```python
t = np.linspace(0, 10, 100)
s = 9.8 * t ** 2 / 2
plt.plot(t, s)
plt.title("Зависимость расстояния от времени", fontsize=20)
plt.xlabel("Время (с)")
plt.ylabel("Расстояние (м)")
plt.show()
```

```python
t = np.linspace(0, 10, 10)
s = 9.8 * t ** 2 / 2
plt.plot(t, s, color="black", marker=".", linewidth=2)
plt.title("Зависимость расстояния от времени", fontsize=20)
plt.xlabel("Время (с)")
plt.ylabel("Расстояние (м)")
plt.show()
```

```python
x = np.linspace(0, 10, 20) ** 2
plt.plot(x, marker=".")
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
y = 1 / x
plt.figure(figsize=(10, 7))
plt.scatter(x, y, marker=".")
plt.show()
```

```python
x = np.linspace(0, 4 * np.pi, 200)
sin = np.sin(x)
sin2 = np.sin(x) ** 2
cos = np.cos(x)
plt.plot(x, sin, label="$sin(x)$", color="red")
plt.plot(x, sin2, label="$sin^2(x)$", color="green")
plt.plot(x, sin * 2, label="$2 \cdot sin(x)$", color="blue")
plt.plot(x, (sin * 2) ** 2, label="$4 \cdot sin^2(x)$", color="black")
# plt.plot(x, cos, label="cos(x)")
plt.legend(loc=1)
plt.show()
```

```python
x = np.linspace(0, 40, 2000)
plt.plot(x, 2 * np.exp(x) * x ** 2, label="$2 \cdot e^x \cdot x^2$", color="red")
plt.plot(x, np.exp(x), label="$e^x$", color="blue")
plt.yscale("log")
plt.legend()
plt.show()
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 200)

figure, axis = plt.subplots(nrows=2, ncols=2)
axis[0, 0].set(title="sin")
axis[0, 0].plot(x, np.sin(x))
axis[0, 1].set(title="cos")
axis[0, 1].plot(x, np.cos(x))
axis[1, 0].set(title="tg")
axis[1, 0].plot(x, np.tan(x))
axis[1, 1].set(title="ctg")
axis[1, 1].plot(x, 1/np.tan(x))
plt.show()

figure.savefig("gr.png")
```

```python
names = ["Василий", "Артем", "Иван", "Зинаида", "Ваня"]
marks_math = [5, 4, 7, 6, 1]
marks_cpp = [10, 6, 3, 5, 9]

plt.bar(names, marks_math, alpha=0.5)
plt.bar(names, marks_cpp, alpha=0.5)
plt.title("Оценки студентов МФТИ", fontsize=20)
plt.xlabel("Студент")
plt.ylabel("Оценка")
plt.show()
```
