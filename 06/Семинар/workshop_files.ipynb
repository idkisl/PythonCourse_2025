### Семинар 6. Работа с файлами

## Задание 1

1. Сгенерируйте текст и сохраните в файл любого формата
2. Напишите функцию, которая принимает на вход имя файла и ключ (число), а затем создает новый файл, где каждый символ исходного файла "сдвинут" на ключ в соответствии с таблицей ASCII (простой шифр Цезаря для всего файла). Напишите вторую функцию для дешифровки.

1. Напишите генератор, принимающую либо строку(путь до файла), либо текстовый объект (from io import TextIOBase), а также список ключевых слов. Генератор должен открывать файл и возвращать строку если в ней есть хотя бы одно ключевое слово.
2. Обработайте все возможные ошибки, которые возникают при работе с файлами.
   

### Задание 3
1. Сгенерируйте json файл
2. Напишите функцию, которая принимает строку json, список полей, которые надо обработать, список ключей, которые надо найти, а также функцию, которая срабатывает, когда в каком-то из полей было найдено ключевое слово.
3. Покройте тестами (Опционально)

Пример:
def parse(json_str: str, required_fields, keywords, keyword_callback):

Подставим json_str = '{"key1" : "word1", "key2" : "word2"}', а required_fields = ["key2"], keywords = ["word2"]. 
Функция keyword_callback будет вызвана только для слова "word2" для ключа "key2"

Распарсить json можно так:
json_doc = json.loads(json_str)
