#### Задание 16 (*)
**Система управления умным домом**

Создайте иерархию классов для управления устройствами в умном доме. Устройства могут быть разными: от простых лампочек до сложных роботов-уборщиков. Каждое устройство должно уметь включаться, выключаться и выполнять свои уникальные функции.

##### Базовые классы:
1. **`Device`** — базовый класс для всех устройств.
   - Методы:
     - `turn_on()` — включает устройство.
     - `turn_off()` — выключает устройство.
     - `get_status()` — возвращает текущее состояние устройства (включено/выключено).

2. **`SmartDevice`** — класс, который наследуется от `Device` и добавляет возможность управления через Wi-Fi.
   - Методы:
     - `connect_to_wifi()` — подключает устройство к Wi-Fi.
     - `send_command(command)` — отправляет команду устройству.

##### Устройства:
3. **`LightBulb`** — класс для умной лампочки, который наследует `SmartDevice`.
   - Методы:
     - `set_brightness(level)` — устанавливает яркость лампочки (от 0 до 100).
     - `change_color(color)` — меняет цвет лампочки.

4. **`Thermostat`** — класс для умного термостата, который наследует `SmartDevice`.
   - Методы:
     - `set_temperature(temp)` — устанавливает температуру.
     - `get_current_temperature()` — возвращает текущую температуру.

5. **`RobotVacuum`** — класс для робота-пылесоса, который наследует `SmartDevice`.
   - Методы:
     - `start_cleaning()` — начинает уборку.
     - `return_to_dock()` — возвращает робота на базу.

##### Сложная часть:
6. **`SmartRoom`** — класс, который управляет всеми устройствами в комнате.
   - Методы:
     - `add_device(device)` — добавляет устройство в комнату.
     - `remove_device(device)` — удаляет устройство из комнаты.
     - `turn_on_all_devices()` — включает все устройства в комнате.
     - `turn_off_all_devices()` — выключает все устройства в комнате.
     - `get_device_statuses()` — возвращает статус всех устройств в комнате.

7. **`AdvancedRobotVacuum`** — класс, который наследует `RobotVacuum` и добавляет новые функции:
   - Методы:
     - `schedule_cleaning(time)` — планирует уборку на определенное время.
     - `avoid_obstacles()` — включает режим избегания препятствий.

##### Задача:
- Создайте класс `SmartHome`, который управляет несколькими комнатами (`SmartRoom`) и устройствами.
  - Методы:
    - `add_room(room)` — добавляет комнату в умный дом.
    - `remove_room(room)` — удаляет комнату из умного дома.
    - `turn_on_all()` — включает все устройства во всех комнатах.
    - `turn_off_all()` — выключает все устройства во всех комнатах.
    - `get_home_status()` — возвращает статус всех устройств во всех комнатах.

---

##### Пример использования:
```python
# Создаем устройства
light = LightBulb()
thermostat = Thermostat()
vacuum = AdvancedRobotVacuum()

# Настраиваем устройства
light.connect_to_wifi()
light.turn_on()
light.set_brightness(75)
light.change_color("blue")

thermostat.connect_to_wifi()
thermostat.set_temperature(22)

vacuum.connect_to_wifi()
vacuum.schedule_cleaning("10:00")

# Создаем комнату и добавляем устройства
living_room = SmartRoom()
living_room.add_device(light)
living_room.add_device(thermostat)
living_room.add_device(vacuum)

# Управляем комнатой
living_room.turn_on_all_devices()
print(living_room.get_device_statuses())

# Создаем умный дом и добавляем комнату
home = SmartHome()
home.add_room(living_room)

# Управляем всем домом
home.turn_on_all()
print(home.get_home_status())
```
