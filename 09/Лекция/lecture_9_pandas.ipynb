**Задание 14.**

Для DataFrame допустимы некоторые методы numpy: напрмимер, транспонирование (.T), ```df.sort_values(by='column', ascending=True/False```)

````python
marks = pd.DataFrame({"Математика": [5, 7, 8], "Программировние": [6, 7, 4], "Философия": [9, 3, 3]})
marks.index = ['Студент 1', 'Студент 2', 'Студент 3']

marks.head()
````

````python
marks.sort_values(by="Программировние", ascending=False)
````

````python
marks["Средний балл"] = (marks["Математика"] + marks["Программировние"] + marks["Философия"]) / 3
marks = marks.sort_values(by="Средний балл", ascending=True)
marks.head()
````

**Задание 15.**

Для чтения данных воспользуйтесь : ```pd.read_csv(...)```
Таблица ```SAT_2010.csv```

Вычислите суммарный балл слушателей и выведите пять лучших студентов. На экран выведите столбцы с их именами и названием школы.

````python
marks = pd.read_csv("SAT_2010.csv")
````

````python
marks.head()
````

````python
marks = marks.loc[:, "School Name":]
marks = marks.dropna()
marks.head()
````

````python
marks.sort_values(by="Number of Test Takers", ascending=False).head()
````

````python
marks["Средний балл"] = (marks["Critical Reading Mean"] + marks["Mathematics Mean"] + marks["Writing Mean"]) / 3
marks.sort_values(by="Средний балл", ascending=False).head()
````
