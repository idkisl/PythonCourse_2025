

## WHERE

```sql
SELECT столбцы таблицы FROM таблицы
    WHERE условие
    
SELECT CompanyName, ContactName, Phone FROM Customer
    WHERE City = "Berlin"  

```

Что может быть условием?

- знаки сравнения (>, <, =)
- логические связки (AND, OR, NOT)
- IN(от, до)
- LIKE "%@mail._ _"           email@mail.ru


```sql
SELECT * FROM Customer
    WHERE Region = (SELECT RegionDescription FROM Region WHERE id = 3)
```
